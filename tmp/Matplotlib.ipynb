```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['font.size'] = 20
```

```python
plt.rcParams['axes.prop_cycle'] = plt.cycler(
    color=['hotpink', '#008000']
)
```

```python
x = np.linspace(0, 1, 10)
y = 2 * np.sin(x)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(
    x, y, 
    linewidth=5,
    label='2 sin(x)',
    zorder=0
    # marker='s', markersize=20,
    # linestyle='-.'
)

ax.plot(
    x, -y, 
    linewidth=5,
    label='-2 sin(x)',
    zorder=0
    # marker='s', markersize=20,
    # linestyle='-'
)


plt.xlabel('X')
plt.ylabel('Y')

plt.legend(loc='center right')


plt.minorticks_on()
plt.grid(which='both', axis='x', zorder=10)

fig.savefig('plot.svg')
```

```python
fig, ax = plt.subplots(figsize=(20, 10))


for shift in np.arange(0, 101):
    plt.plot(
        x + shift/100, y, 
        linewidth=5,
        # marker='s', markersize=20,
        # linestyle='-.'
    )


plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
```

```python
fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(20, 10)
)


ax = axes[0]

ax.plot(
    x, y, 
    linewidth=5,
    color='green'
)

ax.set_xlim(0, 1.1)
ax.set_xticklabels([])

ax.grid()


ax = axes[1]

ax.plot(
    x + 0.1, -2 * y + 1, 
    linewidth=5,
)

ax.set_xlim(0, 1.1)


plt.tight_layout(pad=0.01)
plt.grid()

ax.set_xlabel('X')
```

```python
fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(20, 10)
)


ax.plot(
    x, y, 
    linewidth=5,
    color='green'
)

ax.set_ylabel('Y', color='green')



ax2 = ax.twinx()

ax2.plot(
    x + 0.1, -2 * y + 1,
    linewidth=5,
    color='hotpink'
)

ax2.set_ylabel('Y', color='hotpink')
ax2.tick_params(axis='y', colors='hotpink',)
```

```python
fig, ax = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(20, 10)
)


ax.scatter(
    x, y,
    s=400,
    color='green',
    zorder=10
)

ax.plot(
    [0], [0],
    marker='s',
    markersize=20,
    color='hotpink'
)
```

```python
import os
import warnings

import pandas as pd
import seaborn as sns

warnings.filterwarnings('ignore')

# Graphics in retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
```

```python
DATA_PATH = os.path.join('data', 'beauty.csv')
```

```python
pd.read_csv(DATA_PATH, nrows=2)
```

```python
df = pd.read_csv(DATA_PATH)

df.head()
```

### Examples

```python
obj = df['wage'].hist(figsize=(6, 4), bins=100)
```

```python
_ = sns.distplot(df['wage'], bins=10)
```

```python
_ = df[['wage', 'exper']].hist(figsize=(10, 4), bins=20)
```

```python
_ = sns.countplot(x='female', data=df)
```

```python
_ = sns.countplot(y='female', data=df)
```

```python
_ = sns.countplot(x='female', hue='looks', data=df)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

sns.countplot(x='female', data=df, ax=axes[0])
sns.countplot(x='looks', data=df, ax=axes[1])
```

```python
df.groupby('female')['wage'].count()
```

```python
type(df.groupby('female')['wage'].count())
```

```python
female_counts = df.groupby('female')['wage'].count()

sex_id_to_name = {0: 'male', 1: 'female'}

_ = plt.pie(
    female_counts,
    labels=[sex_id_to_name[i] for i in female_counts.index]
)
```

```python
_ = plt.subplots(figsize=(15, 5))

sns.boxplot(x='wage', data=df)
```

```python
_, axes = plt.subplots(1, 2, sharey=True, figsize=(6, 6))

sns.boxplot(data=df['wage'], ax=axes[0])
sns.violinplot(data=df['wage'], ax=axes[1])
```
