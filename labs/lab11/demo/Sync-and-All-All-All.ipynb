* В первом ответе [Multiprocessing: use only the physical cores?](https://stackoverflow.com/questions/40217873/multiprocessing-use-only-the-physical-cores/40220912) подробно объясняется, как из Питона узнать, сколько на машине ядер (физических и логических):

```python
import psutil

psutil.cpu_count(logical=False)
psutil.cpu_count(logical=True)
```

В целом пост очень рястянутый и немного плохо структурированный, но пара неплохих идей приводятся.

Упоминается библиотека `joblib` как альтернатива `multiprocessing`...
Также говорится, что и в случае `multiprocessing` больше процессоров — не всегда лучше.
Всё зависит от задачи, которая распараллеливается.
Отмечается, что в Питоне "можно" использовать и потоки для CPU-bound задач, но только если, например, написать функции на C/C++ и вызывать их из Питона ("and everytime you do so, one kitten atrociously dies somewhere in the world")...
Даётся совет никогда не использовать для рассчёта задач через `multiprocessing` все доступные ядра, потому что это может негативно сказаться на работе всего компьютера в целом (какая-то работа всегда должна выполняться на машине в фоне).
"Каждый конкретный случай надо рассматривать в рамках решаемой задачи, имплементации решения и доступного hardware".
"Всегда, всегда надо тестировать на несложных задачах".
